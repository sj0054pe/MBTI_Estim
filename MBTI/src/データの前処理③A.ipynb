```python
keys = ['建築家', '論理学者', '指揮官', '討論者', '提唱者', '仲介者', '主人公', '広報運動家', '管理者', '擁護者', '幹部', '領事官', '巨匠', '冒険家', '起業家', 'エンターテイナー']
values = ['INTJ', 'INTP', 'ENTJ', 'ENTP', 'INFJ', 'INFP', 'ENFJ', 'ENFP', 'ISTJ', 'ISFJ', 'ESTJ', 'ESFJ', 'ISTP', 'ISFP', 'ESTP', 'ESFP']

dic = {key: val for key, val in zip(keys, values)}

print(dic)

dict_MBTIType=dic
list_MBTItype=values
```

```python
import glob
import pandas as pd
from tqdm.notebook import tqdm 

for element_type in list_MBTItype:
  print(element_type)
  list_csv=glob.glob('/content/drive/MyDrive/研究室/研究_M2/src/Assets/Assets_Output/Tweet_Datasets/%s/*' % element_type)
  #print(list_csv)
  df_type=None
  for element_fileName in tqdm(list_csv):
    df_each = pd.read_csv('%s' % element_fileName, engine='python')
    df_type = pd.concat([df_type, df_each], axis=0)
  df_type.to_csv('/content/drive/MyDrive/研究室/研究_M2/src/Assets/Assets_Output/Tweet_Datasets/v3_%s.csv' % element_type)
```
